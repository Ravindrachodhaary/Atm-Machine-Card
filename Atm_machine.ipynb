```python
import tkinter as tk
from tkinter import messagebox

# Main ATM class to handle operations
class ATM:
    def __init__(self, master):
        self.master = master
        master.title("ATM Machine")
        self.balance = 50000
        self.password = 12345

        # Labels and Entry widgets
        self.label = tk.Label(master, text="Welcome to the ATM")
        self.label.pack()

        self.pin_label = tk.Label(master, text="Enter your PIN:")
        self.pin_label.pack()

        self.pin_entry = tk.Entry(master, show='*')
        self.pin_entry.pack()

        self.submit_button = tk.Button(master, text="Submit", command=self.check_pin)
        self.submit_button.pack()

        self.balance_button = None
        self.withdraw_button = None
        self.deposit_button = None
        self.exit_button = None

    def check_pin(self):
        pin = int(self.pin_entry.get())
        if pin == self.password:
            self.show_menu()
        else:
            messagebox.showerror("Error", "Invalid PIN")

    def show_menu(self):
        self.clear_screen()

        self.label = tk.Label(self.master, text="ATM Menu")
        self.label.pack()

        self.balance_button = tk.Button(self.master, text="Check Balance", command=self.check_balance)
        self.balance_button.pack()

        self.withdraw_button = tk.Button(self.master, text="Withdraw", command=self.withdraw)
        self.withdraw_button.pack()

        self.deposit_button = tk.Button(self.master, text="Deposit", command=self.deposit)
        self.deposit_button.pack()

        self.exit_button = tk.Button(self.master, text="Exit", command=self.master.quit)
        self.exit_button.pack()

    def clear_screen(self):
        for widget in self.master.winfo_children():
            widget.destroy()

    def check_balance(self):
        messagebox.showinfo("Balance", f"Your current balance is: {self.balance}")

    def withdraw(self):
        self.clear_screen()

        self.label = tk.Label(self.master, text="Enter amount to withdraw:")
        self.label.pack()

        self.amount_entry = tk.Entry(self.master)
        self.amount_entry.pack()

        self.submit_button = tk.Button(self.master, text="Submit", command=self.perform_withdraw)
        self.submit_button.pack()

    def perform_withdraw(self):
        amount = int(self.amount_entry.get())
        if amount > self.balance:
            messagebox.showerror("Error", "Insufficient balance")
        else:
            self.balance -= amount
            messagebox.showinfo("Success", f"{amount} withdrawn. Your new balance is {self.balance}")
            self.show_menu()

    def deposit(self):
        self.clear_screen()

        self.label = tk.Label(self.master, text="Enter amount to deposit:")
        self.label.pack()

        self.amount_entry = tk.Entry(self.master)
        self.amount_entry.pack()

        self.submit_button = tk.Button(self.master, text="Submit", command=self.perform_deposit)
        self.submit_button.pack()

    def perform_deposit(self):
        amount = int(self.amount_entry.get())
        self.balance += amount
        messagebox.showinfo("Success", f"{amount} deposited. Your new balance is {self.balance}")
        self.show_menu()

# Run the application
root = tk.Tk()
atm = ATM(root)
root.mainloop()


```
